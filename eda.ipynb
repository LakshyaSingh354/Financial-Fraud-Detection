```python
import dgl

# Replace 'graph.bin' with the path to your binary file
graph = dgl.load_graphs('graph-S-FFSD.bin')[0][0]
```

```python
graph
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Convert DGL graph to NetworkX graph
nx_graph = graph.to_networkx()

# Visualize the graph
plt.figure(figsize=(10, 10))
nx.draw(nx_graph, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')
plt.show()
```
